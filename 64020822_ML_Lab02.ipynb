<h1>Simple Linear Regression</h1>

```python
#import Library
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Create array 
arr1 = np.array([6,16,26,36,46,56,66,76,86,96]).reshape((-1,1))      #reshape to one col
arr2 = np.array([4,23,10,12,22,23,26,35,40,52])
```

```python
print(arr1)
```

```python
print(arr2)
```

```python
#Create Model variable to get return form function LinearRegression
model = LinearRegression().fit(arr1,arr2)
```

```python
#Calculate Score Coefficient of determination 
r_sq = model.score(arr1,arr2)
print("Coefficient of determination : ",r_sq)
```

```python
#Calculate Intercept 
print("Intercept : ",model.intercept_)
```

```python
#calcolate Slop
print("Slop : ",model.coef_)
```

```python
#Prediction arr1 
y_pred = model.predict(arr1)
print("Prediction : ",y_pred)
```

```python
#Plot from arr1 and arr2
plt.scatter(arr1,arr2)
plt.show()
```

```python
#Plot with line
plt.scatter(arr1,arr2)
plt.plot(arr1,y_pred)
plt.show()
```

```python
#Create random data with numpy and plot it with matplotlib
rnstate = np.random.RandomState(1)
x = 10 * rnstate.rand(50)
y = 2 * x - 5 + rnstate.randn(50)
plt.scatter(x,y)
```

```python
#Create a linear regression model based the positioning of the data and Intercept and prediction
model = LinearRegression(fit_intercept = True)
model.fit(x[:,np.newaxis],y)
xfit = np.linspace(0,10,1000)
yfit = model.predict(xfit[:,np.newaxis])
```

```python
#PLot the estimated linear regression line with matplotlib
plt.scatter(x,y)
plt.plot(xfit,yfit)
plt.show()
```
